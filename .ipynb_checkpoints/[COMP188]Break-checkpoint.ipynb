## Part 1

````python
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers

import numpy as np
import pandas as pd
import math

from matplotlib import cm, gridspec
from matplotlib import pyplot as plt
````

````python
%matplotlib inline
df = pd.read_csv('data/Family Income and Expenditure.csv')

import re

if "Total Food Expenditure" in list(df):    
    df = df.drop("Total Food Expenditure", 1)
reg = re.compile('.*(([Ee]xpenditure)|([Ee]xpenses))$')
expenditure_types = [var for var in list(df) if re.match(reg, var)]
df["TotalExpenditure"] = np.sum(df[expenditure_types],1)

reg_number = re.compile('Number.*')
reg_house = re.compile('House .*')
reg_type = re.compile('Type .*')

remove = [var for var in list(df) if re.match(reg_number, var) 
                                  or re.match(reg_house, var) 
                                  or re.match(reg_type, var)]  + expenditure_types

for var in remove:
    if var in list(df):
        df = df.drop(var, 1)

df = df.rename(index=str, columns={var:(re.sub('[\s+]', '', var))for var in list(df)})

df = df.reindex(np.random.permutation(df.index))
df['Bias'] = 1

X = df["HouseholdHeadAge"]
#y = df["TotalExpenditure"]

y = df["HouseholdHeadAge"]**2 + 10 + np.random.normal(scale=300, size=df.shape[0])
````

### Set up for plots

````python
units = [[5, 5],
         [10, 10],
         [5, 5, 5], 
         [10, 10, 10]]
epochs = [40, 50, 60]
````

````python
# Iterator
plot_i = 1
for epoch in epochs:
    for unit in units:
        model = Sequential()
        for i in range(len(unit)): 
            if i == 0:
                model.add(Dense(units=unit[0], activation='relu', input_dim=1))
            else:    
                model.add(Dense(units=unit[i], activation='relu'))
        model.add(Dense(units=1, activation='linear'))
        model.compile(loss='mse',
        #              optimizer=optimizers.SGD(lr=0.0001),
                      optimizer='rmsprop',
                      metrics=['mse'])

        model.fit(X, y, epochs=epoch, batch_size=32)
        
        predictions = model.predict(X)
        
        plt.subplot(len(epochs), len(units), plot_i)
        plt.scatter(X[:100], y[:100])
        plt.scatter(X[:100], predictions[:100], color='red')
        plot_i += 1
````

Most of these lines fit the parabola very well, but some are close to `y=0` curves.

Half of them are within the `[5,5]` architecture, which isn't too surprising as it is the weaker architecture and shouldn't be performing too well, even the one that manages to have a non-zero function is visually a straight line.

It appears the 50 epochs curves are generally the better fits all across the board at which ever architecture.

The `[5,5]` architecture has peaked my interest, and I want to see the performance of the architecture of `[1]` at the same epochs.

````python
# Iterator
plot_i = 1
for epoch in epochs:
    model = Sequential()
    model.add(Dense(units=1, activation='relu', input_dim=1))
    model.add(Dense(units=1, activation='linear'))
    model.compile(loss='mse',
    #              optimizer=optimizers.SGD(lr=0.0001),
                  optimizer='rmsprop',
                  metrics=['mse'])

    model.fit(X, y, epochs=epoch, batch_size=32)

    predictions = model.predict(X)

    plt.subplot(len(epochs), 1, plot_i)
    plt.scatter(X[:100], y[:100])
    plt.scatter(X[:100], predictions[:100], color='red')
    plot_i += 1
````

Strangely enough, an architecture of `[1]` performs better than an architecture of `[5,5]` overall. It's probably down to chance.

**Shape:** (2406, 188)

**Viable dependent variable:** Sum of positive ratings vs negative ratings, e.g. Positive (Funny, Interesting) vs Negative (Boring)

**Viable predictors:** Transcript/Corpus, pronoun usage (he/she/I)

**Comment:** 

In theory, we could see the usage of a few patterns of words to see how they effect viewer engagement.
